```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the image (replace 'input_image.png' with your actual image path)
input_image_path = '/content/photo_2024-11-29_18-01-45 (1).jpg'  # Assuming the file is uploaded
image = cv2.imread(input_image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Display the results
fig, axes = plt.subplots(1, 1, figsize=(20, 5))
axes.imshow(image, cmap='gray')
axes.set_title("Original Image")
axes.axis('off')
plt.show()
```

```python
# Step 1: Apply vertical flip
vertical_flip = cv2.flip(image, 0)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(vertical_flip, cmap='gray')
axes[1].set_title("Vertical Flip")
axes[1].axis('off')
plt.show()
```

```python
# Step 2: Apply horizontal flip
vertical_horizontal_flip = cv2.flip(vertical_flip, 1)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(vertical_flip, cmap='gray')
axes[1].set_title("Vertical Flip")
axes[1].axis('off')

axes[2].imshow(vertical_horizontal_flip, cmap='gray')
axes[2].set_title("Vertical + Horizontal Flip")
axes[2].axis('off')
plt.show()
```

```python
# Step 3: Apply rotation (e.g., 45 degrees)
h, w = vertical_horizontal_flip.shape
center = (w // 2, h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center,-45, 1.0)
rotated_image = cv2.warpAffine(vertical_horizontal_flip, rotation_matrix, (w, h))

fig, axes = plt.subplots(1, 4, figsize=(20, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(vertical_flip, cmap='gray')
axes[1].set_title("Vertical Flip")
axes[1].axis('off')

axes[2].imshow(vertical_horizontal_flip, cmap='gray')
axes[2].set_title("Vertical + Horizontal Flip")
axes[2].axis('off')

axes[3].imshow(rotated_image, cmap='gray')
axes[3].set_title("Rotated Image")
axes[3].axis('off')
plt.show()
```

```python
# Step 4: Apply 30% zooming to the second output
# Zooming involves cropping and resizing
crop_x = int(w * 0.17)
crop_y = int(h * 0.10)
crop_w = int(w * 0.5)
crop_h = int(h * 0.5)
cropped_image = vertical_horizontal_flip[crop_y:crop_y + crop_h, crop_x:crop_x + crop_w]
zoomed_image = cv2.resize(cropped_image, (w, h), interpolation=cv2.INTER_LANCZOS4)

fig, axes = plt.subplots(1, 5, figsize=(25, 4))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(vertical_flip, cmap='gray')
axes[1].set_title("Vertical Flip")
axes[1].axis('off')

axes[2].imshow(vertical_horizontal_flip, cmap='gray')
axes[2].set_title("Vertical + Horizontal Flip")
axes[2].axis('off')

axes[3].imshow(rotated_image, cmap='gray')
axes[3].set_title("Rotated Image")
axes[3].axis('off')

axes[4].imshow(zoomed_image, cmap='gray')
axes[4].set_title("Zoomed Image")
axes[4].axis('off')
plt.show()
```
